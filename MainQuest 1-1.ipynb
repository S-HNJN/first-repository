```python
import random as rd
import datetime as dt
import math as m

class Account:
    user_count = 0
    bank = 'SC은행'
    user_list = {}
    def __init__(self, user_name, balance):
        self.user_name = user_name
        self.balance = balance
        self.account_num = str(rd.randint(0,999)).zfill(3)+'-'+str(rd.randint(0,99)).zfill(2)+'-'+str(rd.randint(0,999999)).zfill(6)
        self.deposit_count = 0
        self.deposit_records = []
        self.withdrawal_records = []
        Account.user_count += 1
        Account.user_list.setdefault(user_name,{'account_num':self.account_num,'balance':self.balance})

    def get_account_num():
        print(f'현재 은행에 {Account.user_count}개의 계좌가 존재합니다.')

    def deposit(self, money):
        while True:
            try:
                if money > 0:
                    self.balance += money
                    self.deposit_count += 1
                    if self.deposit_count < 5:
                        print(f'{money}원 이 입금되었습니다.\n현재 잔액 : {self.balance}')
                    else:
                        self.balance += money
                        print(f'{money}원이 입금되었습니다.\n입금 횟수가 5회가 되어 이자를 지급합니다.\n이자 : {round(self.balance*0.01,3)}\n현재 잔액 : {self.balance*1.01}')
                        self.balance *= 1.01
                        self.deposit_count = 0
                    Account.user_list[self.user_name]['balance']=self.balance
                    return self.deposit_history(money)
                else:
                    money = int(input('입금할 금액을 입력하세요.(최소 1원 이상): '))
                    
            except Exception:
                money = int(input('숫자를 입력해주세요.(최소 1원 이상): '))
                continue

    def withdraw(self, money):
        while True:
            try:
                if self.balance >= money > 0:
                    self.balance -= money
                    print(f'{money}원 이 출금되었습니다.\n현재 잔액 : {self.balance}')
                    Account.user_list[self.user_name]['balance']=self.balance
                    return self.withdrawal_history(money)
                elif self.balance < money and money > 0:
                    money = int(input('출금할 금액이 부족합니다. 다시 입력해주세요: '))
                else:
                    money = int(input('출금할 금액을 입력하세요.(최소 1원 이상): '))
                    
            except Exception:
                money = int(input('숫자를 입력해주세요.(최소 1원 이상): '))
                continue

    def deposit_history(self, money):
        self.deposit_records.append([f'{money}원',dt.datetime.now().strftime("%Y-%m-%d %H:%M:%S")])

    def withdrawal_history(self):
        self.withdrawal_records.append([f'{money}원',dt.datetime.now().strftime("%Y-%m-%d %H:%M:%S")])

    # 해당 이름을 가진 정보를 반환
    def display_info(user_name):
        try:
            print(f'은행이름:{Account.bank}, 예금주:{user_name}, 계좌번호:{Account.user_list[user_name]['account_num']}, 잔고:{Account.user_list[user_name]['balance']}')

        # user_name이 리스트에 없을 경우
        except Exception:
            print('해당 이름의 유저를 찾을 수 없습니다.')

    # 백만원 이상의 고객의 정보를 반환
    def find_VIP():
        VIP_list = False
        for k,v in Account.user_list.items():
            if v['balance'] >= 1000000:
                Account.display_info(k)
                VIP_list = True
        if not VIP_list:
            print('조건에 해당하는 VIP가 없습니다.')
```
